# I. Количество элементов, больше предыдущего

image.png

```python
my_list = []
while True:
        line = input().strip()
        if line == "0":
            break
        my_list.append(int(line))

n = 0
m = 1
count = 0

while m < len(my_list):
    if my_list[n] < my_list[m]:
        count += 1
    n += 1
    m += 1

print(count)
```

# J. Количество палиндромов

image.png

```python
n = int(input())

count = 0

def reverse_n(n):
    reversed_n = 0
    while n > 0:
        digit = n % 10
        reversed_n = reversed_n * 10 + digit
        n //= 10
    return reversed_n

m = 1

while m <= n:
    if m == reverse_n(m):
        count += 1
    m += 1

print(count)
```
